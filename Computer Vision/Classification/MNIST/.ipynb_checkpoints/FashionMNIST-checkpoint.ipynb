```python
import os
import random
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
import torch
import torchvision
from torchvision.datasets import ImageFolder
from torchvision import transforms, models
from torch.utils.data import DataLoader
```

```python
random.seed(0)
np.random.seed(0)
torch.manual_seed(0)
torch.cuda.manual_seed(0)
torch.backends.cudnn.deterministic = True
```

```python
FashionMNIST_trn = torchvision.datasets.FashionMNIST(os.path.join(os.curdir, 'FashionMNIST_data'), download=True, train=True,
                                                    transform=transforms.Compose([transforms.RandomResizedCrop(20),
                                                                                  transforms.RandomHorizontalFlip(),
                                                                                  transforms.RandomVerticalFlip(),
                                                                                  transforms.ToTensor()]))
FashionMNIST_tst = torchvision.datasets.FashionMNIST(os.path.join(os.curdir, 'FashionMNIST_data'), download=True, train=False,
                                                    transform=transforms.Compose([transforms.ToTensor()]))
```

```python
X_trn = FashionMNIST_trn.data
y_trn = FashionMNIST_trn.targets
```

```python
X_tst = FashionMNIST_tst.data
y_tst = FashionMNIST_tst.targets
```

```python
X_trn = X_trn.unsqueeze(1).float()
X_tst = X_tst.unsqueeze(1).float()
```

```python
X_trn.shape
```

```python
class FashionMNISTNet(torch.nn.Module):
    
    def __init__(self):
        super(FashionMNISTNet, self).__init__()
        
        self.conv1 = torch.nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, padding=2)
        self.maxpool1 = torch.nn.MaxPool2d(kernel_size=2, stride=2)
        self.bn1 = torch.nn.BatchNorm2d(num_features=6)
        self.act1 = torch.nn.ReLU()

        self.conv2 = torch.nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, padding=0)
        self.maxpool2 = torch.nn.MaxPool2d(kernel_size=2, stride=2)
        self.bn2 = torch.nn.BatchNorm2d(16)
        self.act2 = torch.nn.ReLU()

        # self.conv3 = torch.nn.Conv2d(in_channels=12, out_channels=24, kernel_size=3, padding=1)
        # self.maxpool3 = torch.nn.MaxPool2d(kernel_size=2, stride=2)
        # self.bn3 = torch.nn.BatchNorm2d(24)
        # self.act3 = torch.nn.ReLU()

        self.fc1 = torch.nn.Linear(5*5*16, 120)
        self.act4 = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(120, 84)
        self.act5 = torch.nn.ReLU()
        self.fc3 = torch.nn.Linear(84, 10)
        self.act6 = torch.nn.Sigmoid()


    def forward(self, x):
        x = self.conv1(x)
        x = self.act1(x)
        x = self.bn1(x)
        x = self.maxpool1(x)

        x = self.conv2(x)
        x = self.act2(x)
        x = self.bn2(x)
        x = self.maxpool2(x)

        # x = self.conv3(x)
        # x = self.act3(x)
        # x = self.bn3(x)
        # x = self.maxpool3(x)

        x = x.view(x.size(0), x.size(1)*x.size(2)*x.size(3))
        x = self.fc1(x)
        x = self.act4(x)
        x = self.fc2(x)
        x = self.act5(x)
        x = self.fc3(x)
        x = self.act6(x)
        
        return x
```

```python
def train(model,
          X_trn, y_trn,
          X_val, y_val,
          loss, optimizer, lr_scheduler,
          batch_size=100,
          epochs=30):
    device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    loss = loss
    optimizer = optimizer
    lr_scheduler = lr_scheduler
    
    val_acc_hist = []
    val_loss_hist = []
    
    X_val = X_val.to(device)
    y_val = y_val.to(device)
    
    for epoch in range(epochs):
        order = np.random.permutation(len(X_trn))
        for start_idx in range(0, len(X_trn), batch_size):
            optimizer.zero_grad()
            model.train()
            
            batch_idx = order[start_idx:start_idx + batch_size]
            
            X_batch = X_trn[batch_idx].to(device)
            y_batch = y_trn[batch_idx].to(device)
            
            preds = model.forward(X_batch)
            
            loss_value = loss(preds, y_batch)
            loss_value.backward()
            
            optimizer.step()
            
        model.eval()
        val_preds = model.forward(X_val)
        val_loss_hist.append(loss(val_preds, y_val).data.cpu())
        accuracy = (val_preds.argmax(dim=1) == y_val).float().mean().data.cpu()
        val_acc_hist.append(accuracy)
        
        print(f'epoch {epoch}: {accuracy:.6f}')
    print('-----------------------------------')
    
    return val_acc_hist, val_loss_hist
```

```python
model = FashionMNISTNet()
loss = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
```

```python
model_acc, model_loss = train(model,
                              X_trn, y_trn,
                              X_tst, y_tst,
                              loss, optimizer,lr_scheduler,
                              batch_size=100, epochs=30)
```
